```python
from sympy import *
import numpy as np
#### 1.1 to 1.3
g = 9.81 ##gravity
h, a, v0, x  = symbols('h, a, v0, x')

f = -0.5*g * x**2 / (v0**2 * cos(a)**2) + x * tan(a) + h

def tj(alpha, velocity, height):
    return -0.5*9.81 * x**2 / (velocity**2 * cos(alpha)**2) + x * tan(alpha) + height
    
h1 = 1.7
alpha = 40/360 * 2* pi
V01 = 8
```

```python
q = tj(alpha, V01, h1)
alpha2 = pi / 5
r = tj(alpha2, V01, h1)
alpha3 = pi/3
s = tj(alpha3, V01, h1)
p = plot(q,r,s, (x, 5, 6), show=False, legend=False)
p[0].line_color = 'r' ## alpha = 40 deg
p[1].line_color = 'g' ## alpha = pi / 5
p[2].line_color = 'b' ## alpha = pi / 3
p.show()
t = plot(q, (x, 0, 10))
```

```python
### 2 - trajectory analysis
### the value of the derivative of a parabola is zero for Xmax
def fr(alpha, velocity, height):
    return solveset(diff(tj(alpha, velocity, height)), x)
test = fr(pi/4,10,4)
test
```

```python
## 2.2
def throw_dist(alpha, velocity, height): 
    r = solve(tj(alpha, velocity, height))
    l = [x for x in r if x > 0]
    return l[0]
f = throw_dist(pi/4, 8, 4)
f
```
